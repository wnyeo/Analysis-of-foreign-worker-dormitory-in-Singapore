```python
import csv

address = []

# Read the first csv file into nested list
with open('addresses.csv', newline='', encoding='utf8') as f:
    reader = csv.reader(f)
    for row in reader:
        address.append(row)

    
address = address[1:]
```

```python
address[:5]
```

```python
name = []
operator = []
add = []
lat_long = []

for row in address:
    name.append(row[0])
    operator.append(row[1])
    add.append(row[2])
    coordinates = [float(row[3]), float(row[4])]
    lat_long.append(coordinates)
    
```

```python
lat_long
```

```python
# print(lat_long[:20])
import folium
from folium.plugins import HeatMap  

map_folium = folium.Map([1.35255,103.82580], zoom_start=11.4)

HeatMap(lat_long, radius = 8, gradient = {0.2:'blue', 0.4:'purple', 0.6:'red', 1.0:'orange'}).add_to(map_folium)
display(map_folium)

map_folium.save("test1.html")
```

```python
## Write your code below
import folium

from folium.plugins import MarkerCluster       

map_folium = folium.Map([1.35255,103.82580], zoom_start=11.4)

marker_cluster = MarkerCluster().add_to(map_folium)

for i in range(0,len(lat_long)):
    
    lat_long_of_one_listing = lat_long[i]   
    pop_display = name[i]
    
    folium.Marker(
        lat_long_of_one_listing,
        tooltip = pop_display,
    ).add_to(marker_cluster)

display(map_folium)
```
