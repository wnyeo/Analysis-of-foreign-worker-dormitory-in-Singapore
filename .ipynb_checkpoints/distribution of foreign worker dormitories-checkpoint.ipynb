```python
import csv

address = []

# Read the csv file into a nested list
with open('addresses.csv', newline='', encoding='utf8') as f:
    reader = csv.reader(f)
    for row in reader:
        address.append(row)

    
address = address[1:] # removing the header
```

```python
address[0]
```

```python
name = []
lat_long = []

for row in address:
    name.append(row[0])
    coordinates = [float(row[3]), float(row[4])]
    lat_long.append(coordinates)
    
```

```python
lat_long[0]
```

```python
import folium    

map1= folium.Map([1.35255,103.82580], zoom_start=11.4)

for i in range(0,len(lat_long)):
    
    lat_long_of_one_listing = lat_long[i]   
    folium.Marker(
        lat_long_of_one_listing,
        tooltip = name[i]
    ).add_to(map1)

display(map1)
map1.save("markermap.html")
```

```python
# Plotting the Heatmap

from folium.plugins import HeatMap  

map2 = folium.Map([1.35255,103.82580], zoom_start=11.4)

HeatMap(lat_long, radius = 8, gradient = {0.2:'blue', 0.4:'purple', 0.6:'red', 1.0:'orange'}).add_to(map2)

display(map2)
map2.save("heatmap.html")
```
